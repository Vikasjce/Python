# Create a simple pie chart with 2 Sectors

```python
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")

slices=[120,80]
labels=['sixty','forty']
colors=['green','yellow']

plt.pie(slices,labels=labels,colors=colors, wedgeprops={'edgecolor':'black'})

plt.title("A simple Pie Chart")
plt.tight_layout()
plt.show()
```

# Added shadow, percentage and starting angle

```python
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0,0,0,0.1,0]

plt.pie(slices,labels=labels,explode=explode,shadow='true',autopct='%1.1f%%',startangle=90,wedgeprops={'edgecolor':'black'})

plt.title("Percentage of Language used")

plt.show()
```
