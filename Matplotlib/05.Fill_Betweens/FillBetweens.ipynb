```python
from matplotlib import pyplot as plt
import pandas as pd

data = pd.read_csv('data.csv')
Age=data['Age']
Python_developers = data['Python']
All_developers = data['All_Devs']


plt.plot(Age,Python_developers,color='k',linestyle='-',label='Python_dev')
plt.plot(Age,All_developers,color='grey',linestyle='--',label='All_dev')

plt.fill_between(Age,Python_developers)

plt.title("Age vs Medial Salary of Dev(USD)")
plt.xlabel("Ages")
plt.ylabel("Median Salaries")
plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt
import pandas as pd

data = pd.read_csv('data.csv')
Age=data['Age']
Python_developers = data['Python']
All_developers = data['All_Devs']


plt.plot(Age,Python_developers,color='k',linestyle='-',label='Python_dev')
plt.plot(Age,All_developers,color='grey',linestyle='--',label='All_dev')


overall_median=57287


plt.fill_between(Age,Python_developers,overall_median,alpha=0.25)

plt.title("Age vs Medial Salary of Dev(USD)")
plt.xlabel("Ages")
plt.ylabel("Median Salaries")
plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt
import pandas as pd

data = pd.read_csv('data.csv')
Age=data['Age']
Python_developers = data['Python']
All_developers = data['All_Devs']


plt.plot(Age,Python_developers,color='k',linestyle='-',label='Python_dev')
plt.plot(Age,All_developers,color='grey',linestyle='--',label='All_dev')


overall_median=57287


plt.fill_between(Age,Python_developers,overall_median,where=(Python_developers>overall_median),color='green',interpolate=True,alpha=0.25)
plt.fill_between(Age,Python_developers,overall_median,where=(Python_developers<=overall_median),interpolate=True,color='red',alpha=0.25)

plt.title("Age vs Medial Salary of Dev(USD)")
plt.xlabel("Ages")
plt.ylabel("Median Salaries")
plt.tight_layout()
plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt
import pandas as pd

data = pd.read_csv('data.csv')
Age=data['Age']
Python_developers = data['Python']
All_developers = data['All_Devs']


plt.plot(Age,Python_developers,color='k',linestyle='-',label='Python_dev')
plt.plot(Age,All_developers,color='grey',linestyle='--',label='All_dev')


overall_median=57287


plt.fill_between(Age,Python_developers,All_developers,where=(Python_developers>All_developers),color='green',interpolate=True,alpha=0.25,label='Above All')
plt.fill_between(Age,Python_developers,All_developers,where=(Python_developers<=All_developers),interpolate=True,color='red',alpha=0.25,label='Below All')

plt.title("Age vs Medial Salary of Dev(USD)")
plt.xlabel("Ages")
plt.ylabel("Median Salaries")
plt.tight_layout()
plt.legend()
plt.savefig("Fillbetween_Python_developers_all_developers.png")
plt.show()
```
