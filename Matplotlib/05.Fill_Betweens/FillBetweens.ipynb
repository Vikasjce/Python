```python
from matplotlib import pyplot as plt
import pandas as pd

data = pd.read_csv('data.csv')
Age=data['Age']
Python_developers = data['Python']
All_developers = data['All_Devs']


plt.plot(Age,Python_developers,color='k',linestyle='-',label='Python_dev')
plt.plot(Age,All_developers,color='grey',linestyle='--',label='Python_dev')

plt.legend()

plt.show()
```
