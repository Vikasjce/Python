# Simple ScatterPlot

```python
from matplotlib import pyplot as plt
import pandas as pd 

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

plt.style.use('seaborn')
plt.scatter(x,y)

plt.title("Random DataSet")
plt.xlabel("x axis")
plt.ylabel("y axis")

plt.show()
```

```python
from matplotlib import pyplot as plt
import pandas as pd 

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

plt.style.use('seaborn')
plt.scatter(x,y,s=100,c='green',linewidth=1,edgecolors='black',alpha='0.5')

plt.title("Random DataSet")
plt.xlabel("x axis")
plt.ylabel("y axis")


plt.show()
```

```python
from matplotlib import pyplot as plt
import pandas as pd 

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
colors=[7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]

plt.style.use('seaborn')

plt.scatter(x,y,c=colors,linewidth=1,edgecolors='black',s=100,cmap='Greens')

plt.colorbar().set_label("Satisfaction")

plt.tight_layout()
plt.show()
```
